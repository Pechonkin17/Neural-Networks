```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

#  
data_1 = pd.read_csv('Dataset/VICI.csv')

#   
data_1 = data_1[['Open', 'High', 'Low', 'Volume', 'Close']]

#     (X_1)    (y_1)
X_1 = data_1[['Open', 'High', 'Low', 'Volume']]
y_1 = data_1['Close']

#     (80%)   (20%) 
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.2, random_state=42)

#    
model_1 = LinearRegression()

#   
model_1.fit(X_train_1, y_train_1)

#    
y_pred_1 = model_1.predict(X_test_1)

#  
mse_1 = mean_squared_error(y_test_1, y_pred_1)
mae_1 = mean_absolute_error(y_test_1, y_pred_1)
r2_1 = r2_score(y_test_1, y_pred_1)

print(f'Mean Squared Error (MSE): {mse_1}')
print(f'Mean Absolute Error (MAE): {mae_1}')
print(f'R-squared: {r2_1}')

#      
plt.figure(figsize=(10, 5))
plt.plot(y_test_1.values, label='Real Values')
plt.plot(y_pred_1, label='Predicted Values')
plt.legend()
plt.title('Real vs Predicted Close Prices')
plt.xlabel('Sample Index')
plt.ylabel('Close Price')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

#  
data2 = pd.read_csv('Dataset/VICI.csv')

#    (  ,    )
data2 = data2[['Open', 'High', 'Low', 'Volume', 'Close']]

#     (X2)    (y2)
X2 = data2[['Open', 'High', 'Low', 'Volume']]
y2 = data2['Close']

#     (80%)   (20%) 
train_size2 = int(0.8 * len(data2))
X_train2, X_test2 = X2[:train_size2], X2[train_size2:]
y_train2, y_test2 = y2[:train_size2], y2[train_size2:]

#   RandomForestRegressor
model2 = RandomForestRegressor(n_estimators=100, random_state=42)

#  
model2.fit(X_train2, y_train2)

#    
y_pred2 = model2.predict(X_test2)

#  
mse2 = mean_squared_error(y_test2, y_pred2)
mae2 = mean_absolute_error(y_test2, y_pred2)
r2_2 = r2_score(y_test2, y_pred2)

print(f'Mean Squared Error (MSE): {mse2}')
print(f'Mean Absolute Error (MAE): {mae2}')
print(f'R-squared: {r2_2}')

#      
plt.figure(figsize=(10, 5))
plt.plot(y_test2.values, label='Real Values')
plt.plot(y_pred2, label='Predicted Values')
plt.legend()
plt.title('Real vs Predicted Close Prices')
plt.xlabel('Sample Index')
plt.ylabel('Close Price')
plt.show()
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.models import Sequential
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import datetime
from sklearn.model_selection import train_test_split

#    
data = pd.read_csv('Dataset/VICI.csv')
data = data[['Open', 'High', 'Low', 'Volume', 'Close']]

#  
feature_scaler = MinMaxScaler(feature_range=(-1, 1))
scaled_features = feature_scaler.fit_transform(data[['Open', 'High', 'Low', 'Volume']])

#   
target_scaler = MinMaxScaler(feature_range=(-1, 1))
scaled_target = target_scaler.fit_transform(data[['Close']])

# '    
scaled_data = np.hstack((scaled_features, scaled_target))

#     LSTM
def create_windowed_dataset(data, window_size, future_days, step=1):
    X = []
    y = []

    for i in range(0, len(data) - window_size - future_days, step):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size + future_days - 1][4])

    return np.array(X), np.array(y)

window_size = 50
future_days = 1
step = 1
X, y = create_windowed_dataset(scaled_data, window_size, future_days, step=step)

#      
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

#     LSTM
model_LSTM = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.5),
    LSTM(units=256),
    Dropout(0.5),
    Dense(units=1, activation="linear")
])
model_LSTM.compile(optimizer='adam', loss='mse')
model_LSTM.fit(X_train, y_train, epochs=1000, batch_size=32, validation_data=(X_test, y_test))

#       
preds = model_LSTM.predict(X_test)

#         
predictions = target_scaler.inverse_transform(preds.reshape(-1, 1))
actuals = target_scaler.inverse_transform(y_test.reshape(-1, 1))

#  
plt.figure(figsize=(10, 5))
plt.plot(actuals, label='Real Values')
plt.plot(predictions, label='Predicted Values')
plt.legend()
plt.title('Real vs Predicted Close Prices')
plt.xlabel('Sample Index')
plt.ylabel('Close Price')
plt.show()
```
